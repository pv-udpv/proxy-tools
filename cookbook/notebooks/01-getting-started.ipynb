## Quick Decision Guide

```python
def choose_tool(need_pivot=False, behind_nat=False, need_obfuscation=False,
                docker_ingress=False, debug_http=False, load_balance=False):
    if need_pivot:        return "ligolo-ng  — TUN-based pivot into internal network"
    if need_obfuscation:  return "v2ray/xray — protocol obfuscation and censorship bypass"
    if behind_nat and not need_obfuscation:
        return "frp (structured) or chisel (quick HTTP tunnel)"
    if docker_ingress:    return "traefik    — Docker labels, auto Let's Encrypt"
    if load_balance:      return "haproxy (high perf TCP/HTTP) or nginx (HTTP)"
    if debug_http:        return "proxyman   — macOS/iOS traffic inspection and scripting"
    return "go-gost    — flexible, supports almost every scenario via config"

print(choose_tool(behind_nat=True))
print(choose_tool(need_pivot=True))
print(choose_tool(docker_ingress=True))
print(choose_tool(debug_http=True))
```

## Prerequisites Check

```python
import shutil, subprocess, sys

checks = {
    "docker":        "docker version --format '{{.Client.Version}}'",
    "docker compose": "docker compose version --short",
    "go":            "go version",
    "python":        f"{sys.executable} --version",
}

for name, cmd in checks.items():
    try:
        out = subprocess.check_output(cmd.split(), stderr=subprocess.STDOUT, text=True).strip()
        print(f"[OK] {name:<16} {out}")
    except Exception:
        print(f"[--] {name:<16} not found")
```
